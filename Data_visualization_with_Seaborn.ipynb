### Statistical Data Visualization using Seaborn Library (Pokemon dataset)
https://seaborn.pydata.org/

```python
# Import specific libraries to read and visualize data
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the CSV file as a DataFrame (df)
df = pd.read_csv('pokemon.csv', encoding='ISO-8859-1')
#df=pd.read_csv("pokemon.csv",encoding='ISO-8859-1', index_col=0)
```

```python
# Visualize the distribution of various Pokemon's characteristics using a boxplot 
sns.boxplot(data=df.iloc[:,4:10], orient="h")
# Save a *.png file with the display
plt.savefig('sns_boxplot.png', dpi=150)
```

```python
# Display no. of observations for each Dragon using Countplot
sns.countplot(x="Type 1",data=df,palette="rainbow")
plt.xticks(rotation=-45) #=-90
# Save a *.png file with the display
plt.savefig('sns_countplot.png', dpi=150)
```

```python
# Display the correlation relationship betwen the Attack and Defense by using
# the KDE(Kernel density estimation) vs Scatterplot overlay
plt.figure(figsize=(10,6))
#sns.relplot(x='Attack', y='Defense', hue='Stage', data=df)
sns.kdeplot(df.Attack, df.Defense)
#sns.kdeplot(df.Attack,df.Defense,shade=False,kde=True)
sns.scatterplot(df['Attack'],df['Defense'], size=0.03, legend=False)
plt.title('Density vs Scatter');
plt.savefig('sns_kdeplot.png', dpi=150)
```

```python
# Visualisation of Data Patterns
sns.heatmap(df.corr(),annot = True)
plt.savefig('sns_heatmap.png', dpi=150)
```

```python
# # Visualization of a Pairplot
data = df[['Attack', 'Defense', 'Speed', 'HP', 'Legendary']]
sns.pairplot(data, hue='Legendary', diag_kind="kde", kind="scatter", palette="husl")
plt.savefig('sns_pairplot.png', dpi=150)
```
